# **Matplotlib tutorial for beginners**

This notebook is all about Matplotlib, the basic data visualization tool of Python programming language. I have covered Matplotlib object hierarchy, various plot types with Matplotlib and customization techniques associated with Matplotlib.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

x=np.linspace(0,10,100)
fig = plt.figure()

plt.plot(x,np.sin(x),'-')
plt.plot(x,np.cos(x),'--')
```

# The following code produces sine and cosine curves using Pyplot API.

```python
# creates a plot figure 
plt.figure()

# creates the first two panels and set current axis
plt.subplot(2,1,1) # (rows, columns, panel number)
plt.plot(x,np.sin(x))

# creates the second of two panels and set current axis
plt.subplot(2,1,2)
plt.plot(x,np.cos(x))
```
